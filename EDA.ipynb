# **Loading Data and Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import pandas as pd

# Read the historic data
historic_data = pd.read_csv('/content/sample_data/historic.csv')
prediction_input_data = pd.read_csv('/content/sample_data/prediction_input.csv')
```

# **Basic Data Exploration**

```python
# Display the first few rows of historic_data
print("First few rows of historic_data:")
print(historic_data.head())

# Display the first few rows of prediction_input
print("\nFirst few rows of prediction_input:")
print(prediction_input_data.head())
```

```python
# Check the shape of historic_data
print("Shape of historic_data:", historic_data.shape)

# Check the shape of prediction_input_data
print("Shape of prediction_input_data:", prediction_input_data.shape)
```

```python
# Check for missing values in historic_data
missing_values_historic = historic_data.isnull().sum()
print("Missing values in historic_data:")
print(missing_values_historic)

# Check for missing values in prediction_input_data
missing_values_prediction = prediction_input_data.isnull().sum()
print("\nMissing values in prediction_input_data:")
print(missing_values_prediction)
```

# **Summary Statistics**

```python
# Calculate summary statistics for numerical columns in historic_data
historic_data_summary = historic_data.describe()

# Display the summary statistics
print("Summary statistics for historic_data:")
print(historic_data_summary)

# Calculate summary statistics for numerical columns in prediction_input_data
prediction_input_summary = prediction_input_data.describe()

# Display the summary statistics
print("\nSummary statistics for prediction_input_data:")
print(prediction_input_summary)
```

```python
# Explore categorical columns in historic_data
print("Distribution of categories in historic_data:")
for column in historic_data.select_dtypes(include='object').columns:
    print("\nColumn:", column)
    print(historic_data[column].value_counts())

# Explore categorical columns in prediction_input_data
print("\nDistribution of categories in prediction_input_data:")
for column in prediction_input_data.select_dtypes(include='object').columns:
    print("\nColumn:", column)
    print(prediction_input_data[column].value_counts())
```

# **Data Visualization**

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot histograms for numerical variables in historic_data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(historic_data['stars'], bins=20, kde=True, color='skyblue')
plt.title('Distribution of Stars in historic_data')
plt.xlabel('Stars')
plt.ylabel('Frequency')

# Plot histograms for numerical variables in prediction_input_data
plt.subplot(1, 2, 2)
sns.histplot(prediction_input_data['stars'], bins=20, kde=True, color='salmon')
plt.title('Distribution of Stars in prediction_input_data')
plt.xlabel('Stars')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Plot box plots for numerical variables in historic_data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.boxplot(y='stars', data=historic_data, color='skyblue')
plt.title('Box Plot of Stars in historic_data')

# Plot box plots for numerical variables in prediction_input_data
plt.subplot(1, 2, 2)
sns.boxplot(y='stars', data=prediction_input_data, color='salmon')
plt.title('Box Plot of Stars in prediction_input_data')

plt.tight_layout()
plt.show()

# Plot violin plots for numerical variables in historic_data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.violinplot(y='stars', data=historic_data, color='skyblue')
plt.title('Violin Plot of Stars in historic_data')

# Plot violin plots for numerical variables in prediction_input_data
plt.subplot(1, 2, 2)
sns.violinplot(y='stars', data=prediction_input_data, color='salmon')
plt.title('Violin Plot of Stars in prediction_input_data')

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for seaborn
sns.set(style="whitegrid")

# Plot count plots for categorical variables in historic_data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='category', data=historic_data, palette='pastel')
plt.title('Distribution of Category in historic_data')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.countplot(x='main_promotion', data=historic_data, palette='pastel')
plt.title('Distribution of Main Promotion in historic_data')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()

# Plot count plots for categorical variables in prediction_input_data
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.countplot(x='category', data=prediction_input_data, palette='pastel')
plt.title('Distribution of Category in prediction_input_data')
plt.xticks(rotation=45)

plt.subplot(1, 2, 2)
sns.countplot(x='main_promotion', data=prediction_input_data, palette='pastel')
plt.title('Distribution of Main Promotion in prediction_input_data')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for seaborn
sns.set(style="whitegrid")

# Scatter plot of 'stars' (numerical) vs 'category' (categorical) in historic_data
plt.figure(figsize=(10, 6))
sns.stripplot(x='category', y='stars', data=historic_data, palette='pastel')
plt.title('Scatter Plot of Stars vs Category in historic_data')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Stars')
plt.show()

# Box plot of 'stars' (numerical) across different levels of 'category' (categorical) in historic_data
plt.figure(figsize=(10, 6))
sns.boxplot(x='category', y='stars', data=historic_data, palette='pastel')
plt.title('Box Plot of Stars across Categories in historic_data')
plt.xticks(rotation=45)
plt.xlabel('Category')
plt.ylabel('Stars')
plt.show()

# Scatter plot of 'stars' (numerical) vs 'main_promotion' (categorical) in historic_data
plt.figure(figsize=(10, 6))
sns.stripplot(x='main_promotion', y='stars', data=historic_data, palette='pastel')
plt.title('Scatter Plot of Stars vs Main Promotion in historic_data')
plt.xticks(rotation=45)
plt.xlabel('Main Promotion')
plt.ylabel('Stars')
plt.show()

# Box plot of 'stars' (numerical) across different levels of 'main_promotion' (categorical) in historic_data
plt.figure(figsize=(10, 6))
sns.boxplot(x='main_promotion', y='stars', data=historic_data, palette='pastel')
plt.title('Box Plot of Stars across Main Promotions in historic_data')
plt.xticks(rotation=45)
plt.xlabel('Main Promotion')
plt.ylabel('Stars')
plt.show()
```

# **Correlation Analysis**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming historic_data contains only numerical columns
# If not, select only the numerical columns from the dataset
numerical_cols = historic_data.select_dtypes(include=['int64', 'float64'])

# Calculate the correlation matrix
correlation_matrix = numerical_cols.corr()

# Visualize the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```
